```python
import numpy as np
import matplotlib.pyplot as plt

from circuitfp.phase_basis import PhaseBasis
from circuitfp.operators import PhaseOperators
```

```python
bandwidth = 1.0
gap = 0.1
num_modes = 1000

basis = PhaseBasis(bandwidth=bandwidth, gap=gap, num_modes=num_modes)
```

# Construct a phas basis

The effective dimensionality of the low-energy Hilbert space is given by
$$
d_\text{eff} = \frac{\pi}{2} \sqrt{\frac{n\Delta}{2\mathcal{B}}},
$$
which is derived in the reference paper.

```python
print(basis)
```

The phase basis is uniformly distributed over $[-\pi,\pi)$.

```python
basis.phase_grid
```

```python
plt.figure()

plt.scatter(
    basis.phase_grid,
    np.zeros_like(basis.phase_grid)
)

plt.yticks([])
plt.xlabel(r"$\phi$")
plt.title("Phase grid")
```

# Construct phase operators

All operators below are represented in the discretised phase space.

```python
operators = PhaseOperators(basis)
```

```python
# Identity operator
Id = operators.identity()

Id
```

```python
phi = operators.phase()

phi
```

```python
plt.figure()
plt.imshow(phi)
plt.colorbar()
plt.title("Phase operator "r"$\phi_{jk}$")
plt.xlabel(r"$j$")
plt.ylabel(r"$k$")
```

```python
print("dimension =", basis.dimension)
print("spacing =", basis.phase_spacing)
```
